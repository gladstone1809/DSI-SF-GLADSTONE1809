## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. 

```python
#QUESTION why do i need the 'mydata =' ?

#using regex to clean up the 'week' column names

import re

# testing the regex criteria
# print re.sub('^[^x]*x', '', 'x1st.week')
# print re.sub('[.]', '_', '1st.week')
# print re.sub('(?<=_)\w+', 'week', '1st_weksdas')

mydata = mydata.rename(columns=lambda x: re.sub('^[^x]*x', '', x)) #replace all the x's on the weeks
mydata = mydata.rename(columns=lambda x: re.sub('(?<=_)\w+', 'week', x)) #ensure all "week" columns are uniformly spelled

mydata.columns
```

```python
#replace "." with underscores

mydata = mydata.rename(columns=lambda x: re.sub('[.]', '_', x))

mydata.head()
```

```python
#rename artist_inverted to artist_name

mydata.rename(columns = {'artist_inverted':'artist_name'}, inplace = True)

mydata.head()
```

```python
#QUESTION: how to replace the column with this? only works if i print

#un-reverse the artist names

artists = mydata['artist_name']

def comma(name):
    name.split(', ')
    last, first = name.split(', ')
    new_name = re.sub(name, first + ' ' + last, name)
    return new_name

for name in artists:
    if "," in name:
        name = comma(name)
        print name
    else:
        print name
```
