##### 11. Create a dictionary with the values for each of the numeric columns


## Step 3: Describe the data

##### 12. Print the min and max of each column

```python
print "Min Rate score: ", min(my_data['Rate']), " Max Rate score: ", max(my_data['Rate'])
print "Min Verbal score: ", min(my_data['Verbal']), " Max Verbal score: ", max(my_data['Verbal'])
print "Min Math score: ", min(my_data['Math']), " Max Math score: ", max(my_data['Math'])
```

##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
from math import sqrt

# Rate
rate_variance = sum([(xi - sum(my_data['Rate']) / len(my_data['Rate']))**2 for xi in my_data['Rate']]) / len(my_data['Rate'])
rate_stddev = sqrt(rate_variance)

# Verbal
verbal_variance = sum([(xi - sum(my_data['Verbal']) / len(my_data['Verbal']))**2 for xi in my_data['Verbal']]) / len(my_data['Verbal'])
verbal_stddev = sqrt(verbal_variance)

# Maths
math_variance = sum([(xi - sum(my_data['Math']) / len(my_data['Math']))**2 for xi in my_data['Math']]) / len(my_data['Math'])
math_stddev = sqrt(math_variance)

#Checking results against Numpy

import numpy as np

print "Np Rate Std: ", np.std(my_data['Rate']), " My Rate Std: ", rate_stddev
print "Np Verbal Std: ", np.std(my_data['Verbal']), " My Verbal Std: ", verbal_stddev
print "Np Maths Std: ", np.std(my_data['Math']), " My Maths Std: ", math_stddev
```

##### 14. Using MatPlotLib and PyPlot, plot the distribution of the Rate using histograms.

```python
# Numeric and statistics packages:
import numpy as np
import scipy.stats as stats

# Pandas handles dataset loading and manipulation:
import pandas as pd

# import the plotting functions:
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.plotly as py
import plotly.graph_objs as go

# Initialize plotly notebook mode:
plotly.offline.init_notebook_mode()

# Setup matplotlib to display in notebook:
%matplotlib inline


score_data = pd.read_csv(scores_data_path)
score_data.columns = ['State', 'Rate', 'Verbal', 'Math']

# score_data.head()

state = score_data.State.values
rate = score_data.Rate.values
verbal = score_data.Verbal.values
math = score_data.Math.values

# avg = [sum(e)/len(e) for e in zip(math, verbal)]

# for row in score_data:
#     states.append(row[0])

# # for i in rate:
# #        float(i)

plt.hist(rate, color='hotpink')
plt.title("Histogram for Participation Rate")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

fig = plt.gcf()
```

##### 20. Are there any interesting relationships to note?

1. The verbal and math scores are closely related with a pearson coefficient of 0.9.  This is not really surprising, as presumably bright students are more likely do do well in both tests.

2. High scores seem to be inversly related to average participation rates.  It could be that in places where the SAT or uni applications are less common for students, that only the most ambitious and bright students end up sitting the exam.
